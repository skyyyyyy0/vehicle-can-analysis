```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

confusion_matrix = pd.DataFrame({
    "actual_state": ["Moving", "Moving", "Idle", "Idle"],
    "predicted_state": ["Moving", "Idle", "Moving", "Idle"],
    "count": [4, 2, 0, 9]
})

output_path = (
    PROJECT_ROOT
    / "outputs"
    / "validation"
    / "validation_confusion_matrix.csv"
)

output_path.parent.mkdir(parents=True, exist_ok=True)
confusion_matrix.to_csv(output_path, index=False)

print("Saved: outputs/validation/validation_confusion_matrix.csv")
display(confusion_matrix)
```
